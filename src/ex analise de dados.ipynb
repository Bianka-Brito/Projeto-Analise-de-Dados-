```python
import pandas as pd
```

```python
dados= pd.read_csv('C:/Users/biankac/Downloads/dados_vendas_.csv', decimal='.')
dados
```

```python
dados.head(5) #visualizar aprenas primeiras linhas 
```

```python
dados.dropna() #remove valores nulos 
```

```python
#renomeando colunas
novos_nomes= {            
    'data_venda': 'venda'
}
dados_renomeados = dados.rename(columns=novos_nomes)
dados_renomeados.dropna()
dados_renomeados.head()
```

```python
dados.loc[0, 'preco_unitario'] = '150.0' #alterar valor em uma linha, index, nome da coluna e alteração 
dados.head()
```

```python
#alterando uma coluna especifica de todos os produtos com mesmo nome
dados.loc[dados['produto'] == 'Jaqueta', 'preco_unitario'] = 150.0 
dados.loc[dados['produto'] == 'Meia', 'preco_unitario'] = 20.0
dados.loc[dados['produto'] == 'Tênis', 'preco_unitario'] = 200.0
dados.loc[dados['produto'] == 'Camiseta', 'preco_unitario'] = 80.0
dados.dropna()
dados.head()
```

```python
#alterando a quatidade do index 3 
dados.loc[3, 'quantidade'] = '3'
dados.head()
```

```python
#precisei converter a quatidade para floeat pois esta em str, assim a soma deu certo 
dados['quantidade']=dados['quantidade'].astype(float)
```

```python
#somando a quatidade + preco_unitario para dar preço total 
dados['valor_total'] = dados['quantidade'] * dados['preco_unitario']
dados.dropna()
dados.head()
```

```python
#total e media das vendas 
total=dados['valor_total'].sum()
print(f'O total de vendas é: {total}')
media= total / len(dados['valor_total'])
print(f'A média de vendas é: {media}')
```

```python
#total de quantidade e total do valor de cada produto vendido
total_produto=dados.groupby('produto')['quantidade'].sum()
print(f'O total de cada produto vendido foi: {total_produto}')
t_prod= dados.groupby('produto')['valor_total'].sum()
print(f'O preço total de cada produto vendido foi: {t_prod}')
```

```python
#média de do valor cada produto vendido 
media_prod= dados.groupby('produto')['preco_unitario'].mean()
print(f'Média do valor de cada produto vendido: {media_prod}')
```
